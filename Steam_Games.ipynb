```python
import json
import codecs
import math
import re
from statistics import mean 
```

```python
def vector_normalize(vector):
    vector_length = 0
    for value in vector:
        vector_length += value**2        
    inversed_vector_length = 1/math.sqrt(vector_length)
    normalized_vector = []
    for value in vector:
        normalized_vector.append(value*inversed_vector_length)
    return normalized_vector
```

```python
with open (path, "r+", encoding="utf-8") as reader:
    csv_reader = csv.reader(reader, delimiter=",")
    row_count = 0
    columns = []
    data = []
    for row in csv_reader:
        if row_count == 0:
            columns = row
        else:
            data.append({
                columns[0]: int(row[0]),
                columns[1]: row[1],
                columns[2]: row[2],
                columns[3]: int(row[3]),
                columns[4]: row[4],
                columns[5]: row[5],
                columns[6]: row[6].split(";"),
                columns[7]: int(row[7]),
                columns[8]: row[8].split(";"),
                columns[9]: row[9].split(";"),
                columns[10]: row[10].split(";"),
                columns[11]: int(row[11]),
                columns[12]: int(row[12]),
                columns[13]: int(row[13]),
                columns[14]: int(row[14]),
                columns[15]: int(row[15]),
                columns[16]: list(map(int, row[16].split("-"))),
                columns[17]: float(row[17])
            })
        row_count += 1         
```

```python
sorted_data = list(filter(lambda x: int(x['release_date'].split("-")[0]) >= 2015, data))

ban = ['Accounting', 'Documentary', 'Tutorial', 'Animation & Modeling', 'Design & Illustration', 
       'Utilities', 'Education', 'Software Training', 'Game Development', 'Video Production',
      'Photo Editing', 'Audio Production', 'Web Publishing']

for row in ban:
    sorted_data = list(filter(lambda x: row not in x['genres'], sorted_data))

sorted_data[0]
```

```python
genres = []
for game_info in sorted_data:
    for tag in game_info['genres']:
        if tag not in genres:
            genres.append(tag)
genres
```

```python
class_by_orders = []

for game in sorted_data:
    genres_vector = []
    for genre in genres:
        genres_vector.append(1) if genre in game['genres'] else genres_vector.append(0)
    
    tmp = [game['owners'], genres_vector]
    
    class_by_orders.append(tmp)
```

```python
class_by_orders
```

```python
owners = []

for owners_count in class_by_orders:
    if owners_count[0] not in owners:
        owners.append(owners_count[0])
owners.sort(key = lambda x: x)
owners
```

```python
normalized_vectors = []

for owners_count in owners:
    vector = [0 for x in range(29)]
    for item in list(filter(lambda x: x[0] == owners_count, class_by_orders)):
        for i in range(len(item[1])):
            vector[i] += item[1][i]
    normalized_vectors.append({"range": owners_count, "vector": vector_normalize(vector)})
    vector = [0 for x in range(29)]
```

```python
normalized_vectors.sort(key = lambda x: x['range'][0])
```

```python
for vector_group in normalized_vectors:
    print(f"{vector_group['range']}")
```

```python
for vector_group in normalized_vectors:
    print(f"\n{vector_group['range']}")  
    count = 0
    for coord in vector_group['vector']:
        if(round(coord/sum(vector_group['vector']), 2) != 0):
            print(f"{round(coord/sum(vector_group['vector']), 2)} - {genres[count]}")
        count += 1
```

```python
for owners_count in owners:
    paid_games = list(filter(lambda x: x['price'] != 0, sorted_data))
    games_by_owners = list(filter(lambda x: x['owners'] == owners_count, paid_games))
    print(f"\n{owners_count}")
    for genre in genres:
        games_by_genres = list(filter(lambda x: genre in x['genres'], games_by_owners))
        if(len(games_by_genres) != 0):
            mean_income = sum(list(map(lambda y: y['price'] * mean(y['owners']), games_by_genres)))/len(games_by_genres)
            mean_price = sum(list(map(lambda y: y['price'], games_by_genres)))/len(games_by_genres)
            print(f'{genre} - {round(mean_income*0.7)} - {round(mean_price, 2)}')
```

```python
for genre in genres:
    tmp = list(filter(lambda x: genre in x['genres'], sorted_data))
    tmp.sort(key = lambda x: x['owners'][0], reverse=True)
    print(f"{genre}:")
    for i in range(10):
        print(f"{i+1}: {tmp[i]['name']}")
    print("\n")  
```
